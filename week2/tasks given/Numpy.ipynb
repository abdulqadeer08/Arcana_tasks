# Numpy

```python
import numpy as np
```

**Array Creation**

```python
# Create an array from a list
arr = np.array([1, 2, 3, 4])
print("Array:", arr)
```

```python
# Create an array filled with zeros
zeros_array = np.zeros((2, 3))
print("Zeros Array:\n", zeros_array)
```

```python
# Create an array filled with ones
ones_array = np.ones((2, 3))
print("Ones Array:\n", ones_array)
```

```python
# Create an uninitialized array
empty_array = np.empty((2, 3))
print("Empty Array:\n", empty_array)
```

```python
# Create an array with a range of values
range_array = np.arange(5)
print("Range Array:", range_array)
```

```python
# Create an array with evenly spaced values
linspace_array = np.linspace(0, 1, 5)
print("Linspace Array:", linspace_array)
```

```python
# Create an array with values spaced evenly on a log scale
logspace_array = np.logspace(0, 2, 3)
print("Logspace Array:", logspace_array)
```

```python
# Create coordinate matrices from coordinate vectors
x = np.array([1, 2, 3])
y = np.array([4, 5])
X, Y = np.meshgrid(x, y)
print("X:\n", X)
print("Y:\n", Y)

# Stack coordinates into pairs
print("Coordinates List")
coords = np.array([X.ravel(), Y.ravel()]).T
print(coords)

#or in tuples
print("Coordinates tuples")
coordinate_pairs = list(zip(X.ravel().tolist(), Y.ravel().tolist()))
print(coordinate_pairs)
```

```python
# Create an identity matrix
identity_matrix = np.eye(3)
print("Identity Matrix:\n", identity_matrix)
```

```python
# Create a diagonal matrix
diag_matrix = np.diag([1, 2, 3])
print("Diagonal Matrix:\n", diag_matrix)
```

**List Vs Array**

```python
# Python list
lst = [1, 2, 3, 4]
lst_squared = [x**2 for x in lst]
print("List squared:", lst_squared)

# NumPy array
arr = np.array([1, 2, 3, 4])
arr_squared = arr**2
print("Array squared:", arr_squared)
```

**Array Manipulation**

```python
# Reshape an array
arr = np.arange(6)
reshaped_array = arr.reshape((2, 3))
print("Reshaped Array:\n", reshaped_array)
```

```python
# Flatten an array
flattened_array = reshaped_array.ravel()
print("Flattened Array:", flattened_array)
```

```python
# Transpose an array
transposed_array = reshaped_array.T
print("Transposed Array:\n", transposed_array)
```

```python
# Swap two axes of an array
swapped_axes_array = np.swapaxes(reshaped_array, 0, 1)
print("Swapped Axes Array:\n", swapped_axes_array)
```

```python
# Concatenate arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
concatenated_array = np.concatenate((arr1, arr2))
print("Concatenated Array:", concatenated_array)
```

```python
# Split an array into multiple sub-arrays
split_array = np.split(concatenated_array, 2)
print("Split Array:", split_array)
```

```python
# Stack arrays horizontally
hstack_array = np.hstack((arr1, arr2))
print("Horizontal Stack Array:", hstack_array)
```

```python
a1=np.array([[1,2,3],[4,5,6]])
a2=np.array([7,8])
print(a1.shape)
print(a2.shape)
horizontal_stacking=np.hstack((a1,a2.reshape(2,1)))
print(horizontal_stacking)
print(horizontal_stacking.shape)
```

```python
# Stack arrays vertically
vstack_array = np.vstack((arr1, arr2))
print("Vertical Stack Array:\n", vstack_array)
```

```python
# Repeat elements of an array
repeated_array = np.repeat(arr1, 2)
print("Repeated Array:", repeated_array)
```

```python
# Repeat array in a 2D grid
tiled_array_2d = np.tile([[1, 2], [3, 4]], (2, 3))
print("Tiled Array 2D:\n", tiled_array_2d)
```

**Broadcasting**

When performing operations between arrays:

Compare shapes element by element from right to left.

Dimensions must be either:

Equal, OR

One of them is 1 (so it can be stretched).

If neither condition holds → error.

```python
arr = np.array([10, 20, 30])
result = arr + 5
print(result)
```

```python
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])

result = a + b
print(result)
```

```python
#example
sales = np.array([[100, 200, 300],
                  [400, 500, 600],
                  [700, 800, 900]])

discount = np.array([10, 20, 30])  # shape (3,)

# Broadcasts discount across each row
new_sales = sales - discount
print(new_sales)
```

**Linear algebra**

```python
import numpy as np

# Create a matrix from a list of lists
matrix = np.array([[1, 2], [3, 4]])
print("Matrix:\n", matrix)
```

```python
# Transpose a matrix
transposed_matrix = matrix.T
print("Transposed Matrix:\n", transposed_matrix)
```

```python
# Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[2, 0], [1, 3]])
product_matrix = np.dot(matrix1, matrix2)
print("Matrix Product:\n", product_matrix)
```

```python
elementwise_product = matrix1 * matrix2
print("Element-wise Product:\n", elementwise_product)
```

```python
# Compute the inverse of a matrix
inverse_matrix = np.linalg.inv(matrix)
print("Inverse Matrix:\n", inverse_matrix)
```

```python
# Compute the determinant of a matrix
determinant = np.linalg.det(matrix)
print("Determinant:", determinant)
```

```python
# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Coefficient matrix
A = np.array([[2, 3], [4, -1]])

# Right-hand side vector
b = np.array([13, 3])

# Solve the system of linear equations
solution = np.linalg.solve(A, b)
print("Solution (x, y):", solution)
```

**Calculus**

```python
import numpy as np

# Define a function
def f(x):
    return x**2 + 2*x + 1

# Define a point at which to compute the derivative
x = 2.0

# Define a small step size
h = 1e-5

# Compute the derivative using the finite difference method
df_dx = (f(x + h) - f(x - h)) / (2 * h)
print("Numerical Derivative:", df_dx)
```

**Statistical Analayis using NumPy**

```python
import numpy as np

# Sample data
data = [5, 7, 8, 5, 9, 10, 6, 7, 8, 9]

# Convert to a numpy array for convenience
data_array = np.array(data)

# Mean
mean = np.mean(data_array)

# Median
median = np.median(data_array)

# Mode (Note: numpy doesn't have a mode function, so we'll use np.unique for a simple approach)
values, counts = np.unique(data_array, return_counts=True)
mode = values[np.argmax(counts)]

# Range
data_range = np.max(data_array) - np.min(data_array)

# Variance
variance = np.var(data_array, ddof=0)  # Population variance

# Standard Deviation
std_dev = np.std(data_array, ddof=0)  # Population standard deviation

# Interquartile Range (IQR)
q75, q25 = np.percentile(data_array, [75, 25])
iqr = q75 - q25

# Skewness
mean_diff = data_array - mean
skewness = np.mean(mean_diff**3) / (np.std(data_array, ddof=0)**3)

# Kurtosis
kurtosis = np.mean(mean_diff**4) / (np.std(data_array, ddof=0)**4) - 3

# Print results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Range: {data_range}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"IQR: {iqr}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
```
