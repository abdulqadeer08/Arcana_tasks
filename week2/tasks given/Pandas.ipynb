> **Creating data using pandas series**


*   A Series is a one-dimensional labeled array that can store any data type (integers, floats, strings, objects, etc.).
*   Indexing → Each element has a label (index). You can access data by position (like a NumPy array) or by label (like a dictionary).
*   Data Creation → A Series can be created from lists, arrays, dictionaries, scalars, or another Series.
*   Supports vectorized operations (like NumPy) and label-based alignment (like dicts).
*   A Series can have a name (useful in DataFrames), a dtype (data type), and a custom index.

```
pd.Series(
    data=None,     # list, dict, array, scalar, or Series
    index=None,    # custom labels (default: RangeIndex)
    dtype=None,    # specify data type
    name=None,     # optional name for Series
    copy=False,    # deep copy of data if True
    fastpath=False # internal use only
)
```












````python
#from list
age=[30,40,50,25]
age_series = pd.Series(age)
print(age_series)
print(f'Type Of Data In Pandas Series is {age_series.dtype}')
print(f'Index Of Pandas Series is {age_series.index}')
print(f'Values Of Pandas Series is {age_series.values}')
````

````python
#from dictionary
names_scores={"Arooba":80, "Urooj":80, "Rahim":70,"Azib":60}
score_series=pd.Series(names_scores)
print(score_series)
print(f'Type Of Data In Pandas Series is {score_series.dtype}')
print(f'Index Of Pandas Series is {score_series.index}')
print(f'Values Of Pandas Series is {score_series.values}')
````

````python
#from numpy array
arr_scores = np.array([80, 80,70, 60])
arr_series = pd.Series(arr_scores, index=['Arooba', 'Urooj', 'Rahim', 'Azib'])
print(arr_series)
print(f'Type Of Data In Pandas Series is {arr_series.dtype}')
print(f'Index Of Pandas Series is {arr_series.index}')
print(f'Values Of Pandas Series is {arr_series.values}')
````

````python
#handling copy of data

arr = np.array([1, 2, 3])

s1 = pd.Series(arr, copy=False)  # No forced copy
s2 = pd.Series(arr, copy=True)   # Always copy

arr[0] = 99

print("s1:", s1)  # Will reflect the change (99) because it references arr



print("s2:", s2)  # Will not change, independent copy
````

> **Create Data Using Pandas DataFrame**

*   Two-dimensional labeled data structure. A DataFrame is like a table or spreadsheet with rows and columns, where each column can have a different data type (int, float, string, etc.).
*   Flexible data sources. Can be created from lists, dictionaries, Series, NumPy arrays, another DataFrame, or external data sources (CSV, Excel, SQL, JSON, etc.).
*   Indexing and column labels. Each row has an index (row label) and each column has a name (column label).Supports label-based (.loc) and position-based (.iloc) indexing.
*  Heterogeneous data types.Different columns can hold different types of data. For example, one column can be integers, another floats, and another strings.
*   Operations & metadata. Supports vectorized operations, alignment by index, aggregation, merging, joining, and more.Each DataFrame has metadata: index, columns, dtypes, and optional name for columns.

```
pd.DataFrame(
    data=None,        # array, dict, Series, or DataFrame
    index=None,       # row labels
    columns=None,     # column names
    dtype=None,       # data type for all columns (optional)
    copy=False        # deep copy of data if True
)

```






````python
# From List
data = [
    ['Alice', 24, 'New York', 85.5],
    ['Bob', 27, 'Los Angeles', 90.3],
    ['Charlie', 22, 'Chicago', 78.9],
    ['David', 32, 'Houston', 92.0],
    ['Eve', 29, 'Phoenix', 88.7]
]

# Create DataFrame
df = pd.DataFrame(data, columns=['Name', 'Age', 'City', 'Score'])

# Display DataFrame
print(df)
````

````python
#create daily dates from specified date using pandas date range method
daily = pd.DataFrame({
    "date": pd.date_range("2025-01-01", periods=5, freq="D"),
    "sales": [100, 200, 150, 300, 250]
})
print(daily)
````

````python
#create yearly dates from specified date using pandas date range method
yearly = pd.DataFrame({
    "date": pd.date_range("2025-01-01", periods=5, freq="YE"),
    "sales": [100, 200, 150, 300, 250]
})
print(yearly)
````

# **Data Manipulation**

> **Indexing**

*  Series and DataFrame in pandas has an index (row labels).
*  Indexes help in data selection, alignment, reshaping, and joining.
*  Types of indexing:
1.   Default Indexing (0,1,2…)
2.   Custom Indexing (strings, dates, etc.)
3.   Hierarchical (MultiIndex)












````python
# Create dataframe with default index
data = [
    {'Region':'Pakistan','City': 'Islamabad','Sales': 1000,'Profit': 100},
    {'Region':'USA','City': 'California','Sales': 1500,'Profit': 200},
    {'Region':'UK','City': 'London','Sales': 1200,'Profit': 150},
    {'Region':'Australia','City': 'Sydney','Sales':1500,'Profit': 250},
    {'Region':'Pakistan','City': 'Karachi','Sales': 1800,'Profit': 300},
    {'Region':'USA','City': 'New York','Sales': 2000,'Profit': 350},
    {'Region':'UK','City': 'Manchester','Sales': 1700,'Profit': 280},
]
default_index=pd.DataFrame(data)
print(default_index)
````

````python
#change index to custom index
custom_index=default_index.set_index('Region')
print(custom_index)
````

````python
#change index to Hierarchical Index
hierarchical_index=default_index.set_index(['Region','City']).sort_index(level=['Region','City'])
print(hierarchical_index)
````

> **Aggregation**




````python
#group by region
region_grouped=default_index.groupby('Region').sum()
print(region_grouped)
````

````python
# Sum of Sales By region
region_sums = default_index.groupby('Region')['Sales'].sum()
print(region_sums)
````

````python
# Apply multiple aggregation functions
region_agg = default_index.groupby('Region').agg(averageSales=('Sales', 'mean'),
                                                  totalSales=('Sales', 'sum'),
                                                  maxProfit=('Profit', 'max'))
print(region_agg)
````

````python
#Apply aggregation for region and city
region_city_agg = default_index.groupby(['Region', 'City']).agg(averageSales=('Sales', 'mean'),
                                                                totalSales=('Sales', 'sum'),
                                                                maxProfit=('Profit', 'max'))
print(region_city_agg)
````

````python
# Custom function to calculate profit range (max - min)
def profit_range(group):
    return group['Profit'].max()-group['Profit'].min()

# Apply the custom function
range_result = default_index.groupby('Region').apply(profit_range)
print(range_result)
````

````python
# Lambda function to calculate profit range (max - min)
profit_range=default_index.groupby('Region').agg({'Profit': lambda x: x.max()-x.min()})
print(profit_range)
````

````python
# Filter out regions where the total sales are less than 400
filtered_df = default_index.groupby('Region').filter(lambda x: x['Sales'].sum() > 400)
print(filtered_df)
````

````python
# Create pivot table with Region as index, City as columns
pivot = default_index.pivot_table(values='Sales', index='City', columns='Region', aggfunc='sum', fill_value=0)
print(pivot)
````

# **SQL Like Queries Using Pandas**



````python
data1 = {
    'id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'age': [24, 27, 22, 32, 29],
    'salary': [50000, 60000, 45000, 70000, 65000]
}

data2 = {
    'id': [1, 2, 3, 6],
    'dept': ['HR', 'Engineering', 'Sales', 'Marketing']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print(df2)
````

````python
#SELECT column1, column2 FROM table
df1[['name', 'age','salary']]
````

````python
#SELECT * FROM table WHERE condition
df1.loc[df1['salary']>=50000,'name']
````

````python
#SELECT * FROM table WHERE condition1 AND condition2
#SELECT * FROM table WHERE condition1 OR condition2
print(df1[(df1['age'] == 24) & (df1['salary'] == 50000)])
print(df1[(df1['age'] >=24) | (df1['salary'] >= 50000)])
````

````python
#SELECT * FROM table ORDER BY column ASC|DESC
df1.sort_values(by='age', ascending=True)  # Use ascending=False for DESC
````

````python
combine_data=pd.merge(df1,df2,on='id',how='inner')
````

````python
combine_data
````

````python
#SELECT column, COUNT(*) FROM table GROUP BY column
combine_data.groupby('dept')['salary'].sum()
````
